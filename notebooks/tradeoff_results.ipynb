# Tradeoff between flexibility and retention

```python
from plot_utils import load_results, get_tradeoff_dfs, plot_tradeoff
```

## Ablate attention with mask

Category-learning task

```python
path = f'../../results/cat_tradeoff_mask'
ab_levels = ['0p0', '0p1', '0p2', '0p3', '0p4', '0p5', 
             '0p6', '0p7', '0p8', '0p9', '1p0']
fns = [f'{path}_ab_level{ab}' for ab in ab_levels]

cat_results = load_results(ab_levels, fns, rot_cond='unrotated')

cat_dfs = get_tradeoff_dfs(cat_results, 'unrotated', 
                           test_every=20, steps_per_block=2000)
cat_gen_df, cat_ret_df, cat_loss_df = cat_dfs
```

```python
# Parameters for plotting
cat_ab_level_train_max = 1.0  # Maximum attention dropout for retention
cat_last_step = max(cat_ret_df['step'].unique())
cat_xlim = (0, 400)       # X-axis limits for category task
cat_ylim = (0.4, 1.05)    # Y-axis limits for category task
cat_loss_ylim = (0, 1.0)  # Y-axis limits for loss
ab_label = '$p_{a}$'

plot_tradeoff(cat_gen_df, cat_ret_df, cat_loss_df,
              ab_level_train_max=cat_ab_level_train_max,
              last_step=cat_last_step,
              xlim=cat_xlim, ylim=cat_ylim, 
              loss_ylim=cat_loss_ylim, smooth_window=4,
              ab_label=ab_label)
```

Compositional task

```python
path = f'../../results/grid_tradeoff_mask'
ab_levels = ['0p0', '0p1', '0p2', '0p3', '0p4', '0p5', '0p6', '0p7', '0p8']
fns = [f'{path}_ab_level{ab}' for ab in ab_levels]

grid_results = load_results(ab_levels, fns, rot_cond='unrotated')

grid_dfs = get_tradeoff_dfs(grid_results, 'unrotated', 
                            test_every=200, steps_per_block=2000)

grid_gen_df, grid_ret_df, grid_loss_df = grid_dfs
```

```python
# Parameters for plotting
grid_ab_level_train_max = 0.8  # Maximum attention dropout for retention
grid_last_step = max(grid_ret_df['step'].unique())
grid_xlim = (0, grid_last_step) # X-axis limits for category task
grid_ylim = (0.0, 1.05)         # Y-axis limits for category task
grid_loss_ylim = (0, 2.0)       # Y-axis limits for loss
ab_label = '$p_{a}$'

plot_tradeoff(grid_gen_df, grid_ret_df, grid_loss_df,
              ab_level_train_max=grid_ab_level_train_max,
              last_step=grid_last_step,
              xlim=grid_xlim, ylim=grid_ylim, 
              loss_ylim=grid_loss_ylim, smooth_window=20,
              ab_label=ab_label)
```
